```python
'''
实验名称：点亮第一个LED
接线说明：LED模块-->ESP32 IO
         D1-->15

实验现象：程序下载成功后，D1指示灯点亮
注意事项：

'''

#导入Pin模块
from machine import Pin
import time

led1=Pin(15,Pin.OUT)#构建led1对象，GPIO15输出
led1.value(1)#使IO15输出高电平，点亮LED
time.sleep(5)#延时5秒
led1.off()#使led1关闭

"""
构造函数
led = machine.Pin(id, mode,  pull)
构建led对象。id：引脚编号 mode：输入输出方式 pull：上下拉电阻配置

使用方法
led.value([x])
引脚电平值。输出状态：x=0表示低电平，x=1表示高电平；输入状态：无须参数，返回当前引脚值
led.on()
使引脚输出高电平“1”.
led.off()
使引脚输出低电平“0”.
"""
```

```python
'''
实验名称：LED闪烁实验
接线说明：LED模块-->ESP32 IO
         D1-->15

实验现象：程序下载成功后，D1指示灯闪烁
注意事项：

'''


from machine import Pin  #导入Pin模块
import time  #导入time模块

led1=Pin(15,Pin.OUT)#构建led1对象，GPIO15输出

#循环
#写成死循环是为了保持闪烁现象
while True:
    led1.value(1)#使IO15输出高电平，点亮LED
    time.sleep(0.5)#延时0.5秒，表示间隔时长
    led1.value(0)#使IO15输出低电平，熄灭LED
    time.sleep(0.5)#若不加这行代码，led保持熄灭

"""
time.sleep(x)为延时函数，x为延时长度，单位为秒
time.sleep_ms(x)此时单位为毫秒
time.sleep_us(x)此时单位为微秒
单位为秒时允许x为浮点数，单位为毫秒或微秒时x只能为整型

start = time.ticks_ms()可以获取毫秒计时器开始值
delta = time.ticks_diff(time_ticks_ms(), start)用于计算从上电开始到当前时间差值
"""
```

```python
'''
实验名称：LED流水灯实验
接线说明：LED模块-->ESP32 IO
         (D1-D8)-->(15,2,0,4,16,17,5,18)

实验现象：程序下载成功后，LED模块D1-D8指示灯依次点亮后依次熄灭
注意事项：

'''

#导入Pin模块
from machine import Pin
import time

led_pin=[15,2,0,4,16,17,5,18]  #定义LED控制引脚
leds=[]  #定义leds列表，保存LED管脚配置对象
for i in range(8):  #循环8次，0-7
    leds.append(Pin(led_pin[i],Pin.OUT))  #给leds列表添加对象
    #append()方法用于在列表尾部追加元素
# leds=[Pin(led_pin[i],Pin.OUT) for i in range(0,8)]与上面等价


#程序入口
if __name__=="__main__":
    #LED全熄灭
    for n in range(8):
        leds[n].value(0)

    while True:
        #LED逐个点亮
        for n in range(8):
            leds[n].value(1)
            time.sleep(0.05)
        #LED逐个熄灭
        for n in range(8):
            leds[n].value(0)
            time.sleep(0.05)
#先逐个点亮，再逐个熄灭
"""
        #LED逐个点亮
        for n in range(8):
            leds[n].value(1)
            time.sleep(0.05)
        #LED逐个熄灭，反向熄灭
        #reversed()方法使顺序反向
        for n in reversed(range(8)):
            leds[n].value(0)
            time.sleep(0.05)
"""
"""
range()函数是python内置函数，创建一个整数列表，一般用于for循环当中
函数原型：range(start, end, scan)
参数含义：
start：计数从start开始，默认是从0开始，例如range(5)等价于range(0, 5)
end：计数到end结束，但不包括end，例如：range(0, 5)是[0, 1, 2, 3, 4]没有5
scan：每次跳跃的间距，默认为1，例如：range(0, 5)等价于range(0, 5, 1)
"""
```

```python
'''
实验名称：蜂鸣器实验
接线说明：蜂鸣器模块-->ESP32 IO
         (BEEP)-->(25)

实验现象：程序下载成功后，BEEP模块发出声音
注意事项：

'''

#导入Pin模块
from machine import Pin
import time

#定义蜂鸣器控制对象
beep=Pin(25,Pin.OUT)

#程序入口
if __name__=="__main__":
    i=0  #局部变量
    while True:
        i=not i  #非运算，在高低电平之间转换
        beep.value(i)  #控制高低电平输出
        time.sleep_us(250)  #脉冲频率为2KHz, 每个高低电平各占250微秒, 一个周期共500微秒, 故频率为2KHz, 占空比为50%


"""
蜂鸣器分为有源和无源两种
有源蜂鸣器与LED控制逻辑类似，只要供电就会发出声音
无源蜂鸣器控制逻辑则更为复杂，需要一定频率的脉冲才能使其发出声音
无源蜂鸣器最佳发生频率为1.5K-5KHz之间，在其他频率值也能发出声音
改变脉冲信号的频率和占空比可以改变声音的音调和音量

占空比是指在整个周期内高电平所占的时间比重

这里在使用蜂鸣器外设时通过驱动电路相连，之后在使用大功率外设时都需要利用驱动电路
本实验板通过ULN2003芯片控制驱动电路，之后的电机驱动也依靠这块芯片
ULN2003可视为是一个非门电路，即输入高电平，输出低电平；输入低电平，输出高电平
"""
```
